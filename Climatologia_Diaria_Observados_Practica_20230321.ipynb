```python
%matplotlib notebook
import pandas as pd
import matplotlib.pyplot as plt
from datetime import datetime
import numpy as np
```

```python
#abrir datos
prcp = pd.read_excel (r'/Users/arias/OneDrive/Documentos/UCR/TFG/Datos/Climatologia/PRCP_datos_para_practica.xlsx', sheet_name='BD')
```

### Precipitación
Para datos de la forma, columnas= Semana, Día, Mes, Año, Fecha, Estaciones...

```python
#Creamos una copia
prcp1 = prcp.copy()
```

Eliminar todas las columnas que no necesitamos (Semana, Día, Mes, Año)

```python
prcp1.pop('Semana')
prcp1.pop('Día')
prcp1.pop('Mes')
prcp1.pop('Año')
```

```python
#Volver la columna Fecha un datetime
prcp1['Fecha']= pd.to_datetime(prcp1['Fecha'])
prcp1['Fecha']=prcp1['Fecha'].astype(str)
```

Volvemos la columna 'Date' un DateIndex

```python
prcp1.Fecha = pd.to_datetime(prcp1.Fecha)
prcp1.set_index('Fecha',inplace=True)
```

Ocupamos volver todas las columnas a un valor numerico

##### (Se hicieron varios cambios en el excel, por que los datos tenian espacios, notaciones diferentes (PEN,...) o errores (36,,2))

```python
prcp1 = prcp1.apply(pd.to_numeric) #Converting column type to float in Pandas DataFrame
prcp1.dtypes
```

Obtenemos el promedio diario de la climatología

```python
prcp_mean = prcp1.groupby(prcp1.index.strftime("%m%d")).mean() #groupby() se usa agrupando por mes y por día a la vez
```

* Cargar las librerias

```python
import pandas as pd
import glob
import numpy as np
```

## Abrir los archivos de excel

* Archivos diarios que contienen los datos de cada estación

```python
filenames = sorted(glob.glob('20*.csv'))#seleccionar los archivos .csv que inician con 20
```

```python
file = filenames.copy()
```

```python
dataframes = [pd.read_csv(f,header=None) for f in filenames]#crear un dataframe con todos los archivos
```

## Arreglar los datos

Tenemos cuatro tiempos de datos durante el día, se suman los 4 valores para tener el acumulado diario

```python
#Crear una copia para no alterar los datos anteriores
data = dataframes.copy()

#Se suman las 4 columnas de precipitación en cada dataframe
for i, df in enumerate(data):
    data[i] = data[i].iloc[:,1] + data[i].iloc[:,2] + data[i].iloc[:,3] + data[i].iloc[:,4]
```

Transponer los datos para que las columnas sean las estaciones y las filas los días

```python
#Crear una copia para no alterar los datos anteriores
data2 = data.copy()

#Las columnas de cada dataframe se vuelven filas en una tabla
for i, d in enumerate(data2):
    data2[i] = data2[i].T
```

```python
data2Transp = pd.DataFrame(data2) #convertir a dataframe y guardar en una nueva variable
data2Transp
```

* Renombrar las columnas

```python
#Se toma la columna de estaciones del dataframe[0] y se coloca como los nombres de las columnas
data2Transp = data2Transp.rename(dataframes[0].iloc[:,0],axis='columns')
#iloc:selecciona una sola columna
#axis:fila que queremos renombrar
```

```python
data2Transp
```

## Obtener nombres de los archivos

* Obtener lista de archivos

```python
# Utiliza glob para obtener una lista de nombres de archivos
archivos = sorted(glob.glob("20*.csv"))

# Crea una tabla de datos con los nombres de los archivos
df = pd.DataFrame({'Archivo': archivos})
```

* Editar lista de archivos
* Nos interesa que el nombre sea del tipo YYYYMMDD

```python
df['Archivo'] = df['Archivo'].str.replace('.lluvia.csv', '')
```

Agragar el la lista de archivos como fecha en la tabla ya trabajada

```python
data2Transp['Date'] = df['Archivo']
```

* Dar a la fecha formato de datetime

```python
data2Transp['Date'] = pd.to_datetime(data2Transp['Date'])
```

* Volver la fecha un dateindex

```python
data2Transp.Date = pd.to_datetime(data2Transp.Date)
data2Transp.set_index('Date',inplace=True)
```

```python
data2Transp
```

## Obtener los estadisticos necesarios

* Estadisticos: mean, max, min y median

```python
clim_diaria_mean = data2Transp.groupby(data2Transp.index.strftime("%m%d")).mean()#El valor medio pondera los valores observados con la frecuencia con la que se presentan
clim_diaria_max = data2Transp.groupby(data2Transp.index.strftime("%m%d")).max()
clim_diaria_min = data2Transp.groupby(data2Transp.index.strftime("%m%d")).min()
clim_diaria_median = data2Transp.groupby(data2Transp.index.strftime("%m%d")).median()#s más robusta porque es menos sensible a los valores extremos
```

## Guardar los estadisticos en un archivo excel

```python
# Crear el objeto ExcelWriter
writer = pd.ExcelWriter('Climatologia_Diaria_Modelo_20230424.xlsx', engine='xlsxwriter')

# Escribir los DataFrames en diferentes hojas
clim_diaria_mean.to_excel(writer, sheet_name='mean', index=True)
clim_diaria_median.to_excel(writer, sheet_name='median', index=True)
clim_diaria_max.to_excel(writer, sheet_name='max', index=True)
clim_diaria_min.to_excel(writer, sheet_name='min', index=True)

# Guardar el archivo Excel
writer.save()
```
